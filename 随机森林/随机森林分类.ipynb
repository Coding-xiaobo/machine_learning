```python
# 导包
%matplotlib inline
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_wine
```

```python
wine = load_wine()
wine
```

```python
wine.data
```

```python
# 0 1 2 是一个三分类的数据
wine.target
```

### sklearn基本建模流程
1. 实例化模型
2. 训练集带入实例化后的模型进行训练，使用的接口是fit
3. 使用其他接口将测试集导入我们训练好的模型，去获取我们希望获取的结果(score、Ytest)

```python
from sklearn.model_selection import train_test_split
# sklearn的特征和标签要分开导入
Xtrain, Xtest, Ytrain, Ytest = train_test_split(wine.data, wine.target, test_size=0.3)
```

```python
# 1. 实例化
clf = DecisionTreeClassifier(random_state=0) # random_state用于控制树的生成模式 决策树自带随机性 使用random_state参数只能生成一颗决策树
rfc = RandomForestClassifier(random_state=0)
# 2. 训练
clf = clf.fit(Xtrain, Ytrain)
rfc = rfc.fit(Xtrain, Ytrain)
# 3. 获取score
score_c = clf.score(Xtest, Ytest)
score_r = rfc.score(Xtest, Ytest)

print("Single Tree:{}".format(score_c)
      , "Random Forest:{}".format(score_r))
```

```python
# 画出随机森林和决策树在一组交叉验证下的效果对比
# 看在不同的训练集和测试集上模型是否稳定

from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
# n_estimators 是森林中树木的数量，这里随便填了，后面会做n_estimators的学习曲线
rfc = RandomForestClassifier(n_estimators=25)
rfc_s = cross_val_score(rfc, wine.data, wine.target, cv = 10)

clf = DecisionTreeClassifier()
clf_s = cross_val_score(clf, wine.data, wine.target, cv = 10)

plt.plot(range(1,11), rfc_s, label = "RandomForest")
plt.plot(range(1,11), clf_s, label = "Decision Tree")
plt.legend()
plt.show()
```

```python
# 画出在十组(每组10次，cv=10,共100次)交叉验证下的效果对比
# 其实上10次交叉验证和100次交叉验证没区别
rfc_l = []
clf_l = []

for i in range(10):
      rfc = RandomForestClassifier(n_estimators=25)
      rfc_s = cross_val_score(rfc, wine.data, wine.target, cv = 10).mean() # 每组10次交叉验证求均值
      rfc_l.append(rfc_s)
      clf = DecisionTreeClassifier()
      clf_s = cross_val_score(clf, wine.data, wine.target, cv = 10).mean()
      clf_l.append(clf_s)

plt.plot(range(1,11), rfc_l, label = "Random Forest")
plt.plot(range(1,11), clf_l, label = "Decision Tree")
plt.legend()
plt.show()
```

```python
# n_estimators的学习曲线
superpa = []
for i in range(200):
      # 让随机森林在不同的n_estimators下跑了200次
      rfc = RandomForestClassifier(n_estimators= i + 1, n_jobs= -1)
      rfc_s = cross_val_score(rfc, wine.data, wine.target, cv = 10).mean()
      superpa.append(rfc_s)
print(max(superpa), superpa.index(max(superpa)))
plt.figure(figsize=[20,5])
plt.plot(range(1, 201), superpa)
plt.show()
```

```python
# 计算随机森林判断错误的概率
import numpy as np
from scipy.special import comb
# 转换成array进行计算会快很多
np.array([comb(25,i) * (0.2 ** i) * ((1 - 0.2) ** (25 - i)) for i in range(13, 26)]).sum()

# 一棵树的错误率是0.2 而25棵树构成的随机森林错误率是0.000369
```
